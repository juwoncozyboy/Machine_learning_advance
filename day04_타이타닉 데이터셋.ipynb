```python
def encode_features(df):
    features = ['Sex', 'Cabin', 'Embarked']
    for i in features:
        le = LabelEncoder()
        le = le.fit(df[i])
        df[i] = le.transform(df[i])
    return df
titanic_df = encode_features(titanic_df)
```

```python
#원본데이터 재로딩 후 feature 데이터 셋과 label 데이터 셋 추출
titanic_df = pd.read_csv("train.csv")
titanic_df
```

```python
# x , y 데이터 프레임 분리
features = titanic_df.drop('Survived', axis = 1) #피처
labels = titanic_df['Survived']

#
```

```python
features
```

```python
labels
```

```python
#결측치 제거(대체)
features['Age'].fillna(features['Age'].mean(), inplace = True)
features['Cabin'].fillna('N', inplace = True)
features['Embarked'].fillna('N', inplace = True)
```

```python
features = features.drop(['PassengerId','Name','Ticket'], axis = 1)
features
```

```python
features = features.drop('Cabin', axis = 1)
features
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
features['Sex'] = label_encoder.fit_transform(features['Sex'])
features['Embarked'] = label_encoder.fit_transform(features['Embarked'])
features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.5, random_state = 11)
#X_train : 훈련데이터의 피처
#X_test : 테스트데이터의 피처
#y_train : 훈련데이터의 레이블
#y_test : 테스트데이터의 레이블
```

# 모델 학습 및 비교

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dt_clf = DecisionTreeClassifier(random_state = 11)
dt_clf
```

```python
#학습
dt_clf.fit(X_train, y_train)

pred = dt_clf.predict(X_test)
```

```python
print(f"정확도 : {accuracy_score(y_test, pred)}")
```
